```python
from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from urllib.parse import urlparse, parse_qs
import chromedriver_binary
import time
import pandas as pd
```

```python
# race_info_df = pd.read_csv('.csv')
# race_info_df = race_info_df.drop('', axis=1)
```

```python
# オプションを設定してヘッドレスモードを有効にする
chrome_options = Options()
chrome_options.add_argument("--headless")  # GUIをオフに
chrome_options.add_argument("--disable-gpu")  # GPUハードウェアアクセラレーションをオフに。Windows環境では必要な場合があります。
chrome_options.add_argument("--no-sandbox")  # サンドボックスプロセスを無効に。Linux環境の一部のセキュリティ制約を回避するために必要な場合があります。

# ヘッドレスモードでChromeを起動
driver = webdriver.Chrome(options=chrome_options)
# driver = webdriver.Chrome()
```

```python
# 'https://www.boatrace.jp/owpc/pc/race/odds3t?rno=12&jcd=01&hd=20131101'
```

```python
data_list = []
for i in range(1, 13):
    url = 'https://www.boatrace.jp/owpc/pc/race/beforeinfo?rno=12&jcd=02&hd=20231108'
    target_url = f'https://www.boatrace.jp/owpc/pc/race/beforeinfo?rno={i}&jcd' + url.split('&jcd')[-1]
    
    # 日付
    date = url[-8:]
    
    # URLを解析してクエリパラメータを取得
    parsed_url = urlparse(url)
    query_params = parse_qs(parsed_url.query)

    # レース場ID
    place_id = query_params.get('jcd', [None])[0]
    
    # レース番号
    race_number = i

    # スクレイピング対象のURLにアクセス
    driver.get(target_url)

    table_x_path = '/html/body/main/div/div/div/div[2]/div[5]/div[1]/div[1]/table'
    table = driver.find_element(by=By.XPATH, value=table_x_path)
    tbody_elements = table.find_elements(by=By.TAG_NAME, value='tbody')

    for no in range(1, 7):
        d_list = []
        d_list.append(date)
        d_list.append(place_id)
        d_list.append(race_number)
        
        d_list.append(no) # 枠
        
        td_elements = tbody_elements[no-1].find_elements(by=By.TAG_NAME, value='td')        
        name = td_elements[2].find_element(by=By.TAG_NAME, value='a').text
        d_list.append(name)  # ボートレーサー
        
        name_url = td_elements[2].find_element(by=By.TAG_NAME, value='a').get_attribute('href')
        d_list.append(name_url)  # ボートレーサーURL
        
        weight = td_elements[3].text
        d_list.append(weight)  # 体重
                        
        display_time = td_elements[4].text
        d_list.append(display_time)  # 展示タイム
        
        tilt = td_elements[5].text
        d_list.append(tilt)  # チルト
        
        component_replacement = td_elements[7].text
        d_list.append(component_replacement)  # 部品交換
        
        temperature_x_path = '/html/body/main/div/div/div/div[2]/div[5]/div[2]/div[2]/div[1]/div[1]/div/span[2]'
        temperature = driver.find_element(by=By.XPATH, value=temperature_x_path).text
        d_list.append(temperature)  # 気温
        
        weather_x_path = '/html/body/main/div/div/div/div[2]/div[5]/div[2]/div[2]/div[1]/div[2]/div/span'
        weather = driver.find_element(by=By.XPATH, value=weather_x_path).text
        d_list.append(weather)  # 天気
        
        wind_x_path = '/html/body/main/div/div/div/div[2]/div[5]/div[2]/div[2]/div[1]/div[3]/div/span[2]'
        wind_speed = driver.find_element(by=By.XPATH, value=wind_x_path).text
        d_list.append(wind_speed)  # 風速
        
        water_x_path = '/html/body/main/div/div/div/div[2]/div[5]/div[2]/div[2]/div[1]/div[5]/div/span[2]'
        water_temperature = driver.find_element(by=By.XPATH, value=water_x_path).text
        d_list.append(water_temperature)  # 水温
        
        wave_x_path = '/html/body/main/div/div/div/div[2]/div[5]/div[2]/div[2]/div[1]/div[6]/div/span[2]'
        wave_height = driver.find_element(by=By.XPATH, value=wave_x_path).text
        d_list.append(wave_height)  # 波高
        
        
        data_list.append(d_list)
```

```python
columns = [
    'date', 'place_id', 'race_number', 'no', 'name', 'name_url', 'weight', 'display_time', 'tilt',
    'component_replacement', 'temperature', 'weather', 'wind_speed', 'water_temperature', 'wave_height'
]

before_df = pd.DataFrame(data_list, columns=columns)
before_df.head()
```
